```python
re_Rating = r"\b\d\.\d\b"
re_Num_of_Reviews = r"[\d]+\s+Ratings"
re_remarks =r"\·.*"
```

```python
df1["rating"] = df1["rating"].astype(str)
```

```python
ratings_dict = {"Ratings":re_Rating,"No_of_Reviews":re_Num_of_Reviews,"Remarks":re_remarks}

for col_name, pattern in ratings_dict.items():
    df1[col_name] = df1["rating"].apply(lambda x:re.search(pattern,x).group() if re.search(pattern,x) != "nan" else np.NaN)
```

```python
ratings_dict = {"Ratings":re_Rating,"No_of_Reviews":re_Num_of_Reviews,"Remarks":re_remarks}

for col_name, pattern in ratings_dict.items():
    df1[col_name] = df1["rating"].apply(lambda x:re.findall(pattern,x) if re.search(pattern,x) != np.NaN else np.NaN)
```
